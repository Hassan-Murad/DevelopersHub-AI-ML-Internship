```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-darkgrid')
PALETTE = ['#2196F3', '#4CAF50', '#F44336']

print("✅ All imports successful.")
```

## 2. Load the Dataset

```python
# Load Iris dataset directly from seaborn (no download needed)
df = sns.load_dataset('iris')

print(f"✅ Dataset loaded successfully!")
print(f"\n📐 Shape: {df.shape}  ({df.shape[0]} rows × {df.shape[1]} columns)")
print(f"\n📋 Column Names: {list(df.columns)}")
print(f"\n🌸 Species: {df['species'].unique().tolist()}")

print("\n--- First 5 Rows ---")
df.head()
```

## 3. Data Inspection

```python
# Dataset info — dtypes, non-null counts
print("=" * 50)
print("DATASET INFO")
print("=" * 50)
df.info()
```

```python
# Descriptive statistics
print("=" * 50)
print("DESCRIPTIVE STATISTICS")
print("=" * 50)
df.describe().round(2)
```

```python
# Check for missing values
print("Missing Values:")
missing = df.isnull().sum()
print(missing if missing.any() else "  None ✅ — dataset is clean")

print("\nClass Distribution:")
print(df['species'].value_counts())
print("\n✅ Perfectly balanced — 50 samples per species.")
```

## 4. Visualizations

### 4.1 Scatter Plots — Feature Relationships

```python
# Pairplot — all feature combinations coloured by species
pair = sns.pairplot(df, hue='species', palette=PALETTE,
                    diag_kind='kde', plot_kws={'alpha': 0.7, 's': 50})
pair.fig.suptitle('Iris Dataset — Pairwise Feature Relationships', y=1.02,
                   fontsize=14, fontweight='bold')
pair.fig.savefig('scatter_pairplot.png', dpi=150, bbox_inches='tight')
plt.show()

print("Insight: Petal length and petal width clearly separate Setosa from the other two species.")
```

```python
# Focused scatter: Petal Length vs Petal Width (best separator)
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Petal dimensions
for species, color in zip(df['species'].unique(), PALETTE):
    subset = df[df['species'] == species]
    axes[0].scatter(subset['petal_length'], subset['petal_width'],
                    label=species.capitalize(), color=color, alpha=0.8, s=60, edgecolors='white')
axes[0].set_xlabel('Petal Length (cm)', fontsize=11)
axes[0].set_ylabel('Petal Width (cm)', fontsize=11)
axes[0].set_title('Petal Length vs Petal Width', fontweight='bold')
axes[0].legend()

# Sepal dimensions
for species, color in zip(df['species'].unique(), PALETTE):
    subset = df[df['species'] == species]
    axes[1].scatter(subset['sepal_length'], subset['sepal_width'],
                    label=species.capitalize(), color=color, alpha=0.8, s=60, edgecolors='white')
axes[1].set_xlabel('Sepal Length (cm)', fontsize=11)
axes[1].set_ylabel('Sepal Width (cm)', fontsize=11)
axes[1].set_title('Sepal Length vs Sepal Width', fontweight='bold')
axes[1].legend()

plt.suptitle('Iris Dataset — Scatter Plots by Species', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('scatter_focused.png', dpi=150, bbox_inches='tight')
plt.show()
```

### 4.2 Histograms — Value Distributions

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
fig, axes = plt.subplots(2, 2, figsize=(13, 9))
axes = axes.flatten()

for ax, feat in zip(axes, features):
    for species, color in zip(df['species'].unique(), PALETTE):
        subset = df[df['species'] == species][feat]
        ax.hist(subset, bins=15, alpha=0.65, color=color,
                label=species.capitalize(), edgecolor='white')
        ax.axvline(subset.mean(), color=color, linestyle='--', linewidth=1.5)

    ax.set_title(feat.replace('_', ' ').title(), fontweight='bold')
    ax.set_xlabel('cm')
    ax.set_ylabel('Frequency')
    ax.legend(fontsize=9)

plt.suptitle('Iris Dataset — Feature Distributions by Species', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('histograms.png', dpi=150, bbox_inches='tight')
plt.show()

print("Insight: Setosa has distinctly smaller petals. Versicolor and Virginica overlap more.")
```

### 4.3 Box Plots — Outlier Detection

```python
fig, axes = plt.subplots(2, 2, figsize=(13, 9))
axes = axes.flatten()

for ax, feat in zip(axes, features):
    sns.boxplot(
        data=df, x='species', y=feat, ax=ax,
        palette=PALETTE, width=0.5,
        flierprops={'marker': 'o', 'markerfacecolor': 'red',
                    'markersize': 5, 'alpha': 0.7}
    )
    ax.set_title(feat.replace('_', ' ').title(), fontweight='bold')
    ax.set_xlabel('')
    ax.set_ylabel('cm')
    ax.set_xticklabels([s.capitalize() for s in df['species'].unique()])

plt.suptitle('Iris Dataset — Box Plots (Outlier Detection)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('boxplots.png', dpi=150, bbox_inches='tight')
plt.show()

print("Insight: Sepal width shows a few outliers in Setosa. Petal features show clean separation.")
```

### 4.4 Correlation Heatmap

```python
fig, ax = plt.subplots(figsize=(7, 5))

corr = df[features].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            mask=mask, ax=ax, linewidths=0.5,
            cbar_kws={'shrink': 0.8})
ax.set_title('Iris Dataset — Feature Correlation Heatmap', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()

print("Insight: Petal length and petal width are highly correlated (0.96).")
print("Sepal width is weakly or negatively correlated with other features.")
```

### 4.5 Violin Plots — Distribution Shape

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, feat in zip(axes, ['petal_length', 'petal_width']):
    sns.violinplot(data=df, x='species', y=feat, ax=ax,
                   palette=PALETTE, inner='quartile', cut=0)
    ax.set_title(f'{feat.replace("_", " ").title()} by Species', fontweight='bold')
    ax.set_xlabel('')
    ax.set_ylabel('cm')
    ax.set_xticklabels([s.capitalize() for s in df['species'].unique()])

plt.suptitle('Iris Dataset — Violin Plots (Distribution Shape)', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('violin_plots.png', dpi=150, bbox_inches='tight')
plt.show()
```
